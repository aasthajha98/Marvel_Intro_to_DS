## Pivot
A pivot table is a table of grouped values that aggregates the individual items of a more extensive table within one or more discrete categories. 

Here we are pivoting our dataset to see the average power for each alignment type. 

```python
pd.pivot_table(df, 
               index='Alignment_x', 
               values='Power',
               aggfunc='mean')
```

We can also create more complex pivots with multiple columns. Below we are pivoting our dataframe to look at the ability scores of each character. 

```python
pd.pivot_table(df, 
               index='Name', 
               values= ['Height','Weight', 'Intelligence_x','Strength','Speed', 'Durability_x', 'Power', 'Combat'],
               )
```

We can also pivot to see the maximum ability score for each species of characters.  

```python
pd.pivot_table(df, 
               index='Race', 
               values= ['Height','Weight', 'Intelligence_x','Strength','Speed', 'Durability_x', 'Power', 'Combat'],
               aggfunc='max')
```

## Aggregation
We can also aggregate our data along a variable using an aggregate function to group the field by. Below we can see the average intelligence of characters for each alignment. 

```python
# Group by
df.groupby(['Alignment_x'])[['Intelligence_x']].mean().reset_index().sort_values(by = ['Intelligence_x','Alignment_x'], 
                                                                                       ascending = True)
```

Similarly, we can aggregate the average abilities of all of our characters by gender. From this aggregate table we can see that Male characters on average tend to have greater abilities than female or unknown gendered characters. 

```python
df.groupby(['Gender'])[['Height','Weight', 'Intelligence_x','Strength','Speed', 'Durability_x', 'Power', 'Combat']].mean(
    ).reset_index()
```

# Data Visualization


## Pie Chart
With the Pie Chart below we can visualize what proportion of our characters belong to each alignment. As we can see the amount of characters belonging to each alignment seems to be fairly close to one another with their being more unknown and neutral characters than there are good or bad - implying perhaps that good vs bad is a bit harder determine than we would think.

```python
# create data
df_pi = df.groupby(['Alignment_x'])[['Intelligence_x']].mean().reset_index().sort_values(by = 
                                ['Intelligence_x','Alignment_x'], ascending = True).replace(to_replace='-', value='Unkown')
    
team_size=df_pi['Intelligence_x']
teams = df_pi['Alignment_x']

plt.title('Character Alignment', fontsize = 20)
sns.set(style="darkgrid")
# Create a pieplot
plt.pie(team_size,
        labels=teams)
plt.legend()
# add a circle at the center
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)

p.set_size_inches(10,10)
plt.show();
```

## Bar Chart

In the bar chart below, we are using python tools in order to visualize how many null values there are in 5 particular columns. This is helpful during the EDA process to visualize in which ways the data needs to be cleaned. For example for *Combat* and *Total* a best practice would be to replace the null values with 0 since they are quantitative variables. For *Agility, Accelerated Healing* and *Lantern Power Ring* it would be better to replace the nulls with False since they are boolean variables.

```python
fig, ax = plt.subplots(figsize=(10,5))

df[null_columns].isnull().sum()[5:10].plot(kind = 'bar', color=['r','y', 'g', 'b', 'm'])

for rect, label in zip(ax.patches, df[null_columns].isnull().sum()[5:10]):
    ax.text(rect.get_x() + rect.get_width() / 2, rect.get_height() + 0.05, label, ha="center", va="bottom")
    
plt.xlabel('Column Name', fontsize=18)
plt.ylabel('Number of nulls', fontsize=16)
plt.xticks(rotation=0);
```

## Heat Map
Here is a heat map showing how species vary on average in ability. We can use this type of visualization to quickly deduce how one species varies from another based on their abilities. In order to make this heat map we are using the **pivot_table** function we saw above to first create a pivot table that averages the abilities by species. 

```python
ability = ['Height','Weight', 'Intelligence_x','Strength','Speed', 'Durability_x', 'Power', 'Combat']
df_map = pd.pivot_table(df, 
               index='Race', 
               values= ability,
               aggfunc='mean')[1:]
species = df_map.index.to_list()
```

```python
fig, ax1 = plt.subplots(figsize=(10,10))
sns.heatmap(df_map, xticklabels= ability, yticklabels=species,  linewidths=.5, cmap="GnBu_r",  vmin=0, vmax=500, center = 50 )
ax1.set_title('Species vs Ability Heat Map', fontsize=20)
plt.xlabel('Ability', fontsize=16)
plt.ylabel('Species', fontsize=16);
```
